```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line plot

```python
x=[2,5,7,15,20]
y=[35,87,39,10,20]

plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

## Draw a sine wave

```python
x = np.linspace(0,10)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

### Change Plot color and linestyle 

```python
plt.plot(x,y,color="red", marker="o", linestyle="dashed")
```

### Add x-label, y-label, tile

```python
plt.plot(x,y,color="red", marker="o", linestyle="dashed")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Sine Wave")
plt.show()
```

## Draw a cosine wave

```python
x = np.linspace(0,10)
y = np.cos(x)

plt.plot(x,y,color="green", marker="*", linestyle="dashed")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Cosine Wave")
plt.show()
```

## Draw Scatter Plot

### Generate random data points

```python
x = np.random.rand(200)
y = np.random.rand(200)
```

```python
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Scatter Plot")
plt.show()
```

### Change colors of Scatter plot properties

```python
x = np.random.rand(200)
y = np.random.rand(200)

colors=np.random.rand(200)

plt.scatter(x,y,c=colors)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Scatter Plot")
plt.show()
```

### Change  The alpha blending value between 0 (transparent) and 1 (opaque)

```python
x = np.random.rand(200)
y = np.random.rand(200)

colors=np.random.rand(200)
alpha=0.3

plt.scatter(x,y,c=colors,alpha=alpha)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Scatter Plot")
plt.show()
```

```python
x = np.random.rand(200)
y = np.random.rand(200)

colors=np.random.rand(200)
alpha=0.8

plt.scatter(x,y,c=colors,alpha=alpha)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Title: Scatter Plot")
plt.show()
```

### Bar Plot

```python
student = ['Ram','Sham','Jodhu','modhu']
percentage = [67,89,70,29]


plt.bar(student,percentage)
plt.xlabel("student")
plt.ylabel("percentage")
plt.title("Bar plot")
plt.show()
```

### Horizontal Bar plot

```python
student = ['Ram','Sham','Jodhu','modhu']
percentage = [67,89,70,29]


plt.barh(student,percentage)
plt.xlabel("percentage")
plt.ylabel("student")
plt.title("Horizontal Bar plot")
plt.show()
```

## Pie Chart

```python
party=["BJP", "Congress", "AJP", "AAP"]
vote=np.array([40,30,20,10])

plt.pie(vote,labels=party)
plt.show()
```
