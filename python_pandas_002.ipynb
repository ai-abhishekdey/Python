## iloc[ ] and loc[ ]

### Difference 1 :

**iloc** - looks for row index number and column index number

**loc** - looks for row index number but column names

## Difference 2:

**iloc -** row_idx =  start : end+1 , coln_idx = start : end + 1

**loc -** row_idx = start : end

### Syntax

df.iloc[row_idx,[colmn_idx]]

df.loc[row_idx,[colmn_names]]

### Extract records of first 3 rows and first 3 columns using iloc() and loc()

```python
df.iloc[0:4,[0,1,2]]
```

```python
df.loc[0:3,['PassengerId','Survived','Pclass']]
```
